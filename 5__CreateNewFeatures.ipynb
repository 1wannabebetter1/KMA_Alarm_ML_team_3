```python
import datetime
import numpy as np
import pandas as pd
```

```python
INPUT_DATA_FOLDER = "data/merged_dataset"
INPUT_DATA_FILE = "merged.csv"

INPUT_DATA_FILE_ISW = "merged_without_ISW.csv"

OUTPUT_FOLDER = "data/merged_dataset"
OUTPUT_DATA_FILE = "new_features.csv"
```

```python
df = pd.read_csv(f"{INPUT_DATA_FOLDER}/{INPUT_DATA_FILE}", sep=";")
df_no_isw = pd.read_csv(f"{INPUT_DATA_FOLDER}/{INPUT_DATA_FILE_ISW}", sep=";")
```

```python
df.head(5)
```

```python
df_no_isw.head(5)
```

## isAlarm feature to check if it is alarm

```python
df['isAlarm'] = np.where(df.event_region_title.notnull(), True, False)
df_no_isw['isAlarm'] = np.where(df_no_isw.event_region_title.notnull(), True, False)
```

```python
df.head(15)
```

```python
df_no_isw.head(15)
```

## number of regions with alarm

```python
df2=df.query('isAlarm == True').groupby(['hour_datetimeEpoch', 'region']).size().groupby('hour_datetimeEpoch').size().to_frame('df2')
df_no_isw_2 = df_no_isw.query('isAlarm == True').groupby(['hour_datetimeEpoch', 'region']).size().groupby('hour_datetimeEpoch').size().to_frame('df2')
```

```python
df2.head(5)
```

```python
df_no_isw_2.head(5)
```

```python
df.query('hour_datetimeEpoch == 1647154800').head(15)
```

```python
df3 = df.merge(df2, how = 'left', on="hour_datetimeEpoch")
df_no_isw_3 = df_no_isw.merge(df_no_isw_2, how = 'left', on="hour_datetimeEpoch")
df3.head(5)
```

```python
df_no_isw_3.head(5)
```

```python
df3 = df3.rename(columns= {"df2":"regions_with_alert"})
df_no_isw_3 = df_no_isw_3.rename(columns= {"df2":"regions_with_alert"})
```

```python
df3['event_event_time'] = pd.to_datetime(df['event_event_time'])
df3['alerts_last_24h'] = df3.groupby([pd.Grouper(key='event_event_time', freq='24h', origin='start'), 'region']).cumcount() + 1
```

```python
df_no_isw_3['event_event_time'] = pd.to_datetime(df['event_event_time'])
df_no_isw_3['alerts_last_24h'] = df_no_isw_3.groupby([pd.Grouper(key='event_event_time', freq='24h', origin='start'), 'region']).cumcount() + 1
```

```python
df3['alerts_last_24h_all_reg'] = df3.query('event_all_region == 1').groupby([pd.Grouper(key='event_event_time', freq='24h', origin='start'), 'region']).cumcount() + 1
df_no_isw_3['alerts_last_24h_all_reg'] = df_no_isw_3.query('event_all_region == 1').groupby([pd.Grouper(key='event_event_time', freq='24h', origin='start'), 'region']).cumcount() + 1
```

```python
df3.head(5)
```

```python
df3.to_csv(f"{OUTPUT_FOLDER}/{OUTPUT_DATA_FILE}", sep=";", index=False)
```

```python
df_no_isw_3.to_csv(f"{OUTPUT_FOLDER}/new_features_no_ISW.csv", sep=";", index=False)
```
