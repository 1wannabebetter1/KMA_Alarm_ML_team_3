```python
from IPython.display import display, HTML
import pandas as pd
from bs4 import BeautifulSoup
```

```python
INPUT_FOLDER = "isw_raw_pandas"
INPUT_DATA_FILE = "all_isw_new.csv"

OUTPUT_FOLDER = "isw_to_text"
OUTPUT_DATA_FILE = "isw_text.csv"
```

```python
df=pd.read_csv(f"{INPUT_FOLDER}/{INPUT_DATA_FILE}", sep=";")
```

```python
df
```

```python
text_row = df.iloc[13]
page_html_text = text_row["main_html"]
```

```python
display(HTML(page_html_text))
```

```python
def remove_names_and_date(page_html_text):
    if(page_html_text.find("This map is updated daily alongside the static maps present in this report")!=-1):
        return page_html_text[page_html_text.find('This map is updated daily alongside the static maps present in this report')+len("This map is updated daily alongside the static maps present in this report"):]
    if(page_html_text.find("EST\n")!=-1):
        return page_html_text[page_html_text.find('EST\n')+len("EST\n"):]
    elif(page_html_text.find("EST\xa0")!=-1):
        return page_html_text[page_html_text.find('EST\xa0')+len("EST\xa0"):]
    elif(page_html_text.find("ET\n")!=-1):
        return page_html_text[page_html_text.find('ET\n')+len("ET\n"):]
    else:
        return page_html_text[page_html_text.find('ET\xa0')+len("ET\xa0"):]
```

```python
df['main_html_v2'] = df['main_html'].apply(lambda x: remove_names_and_date(x))
```

```python
test_row = df.iloc[300]
main_html_v4 = test_row["main_html_v2"]
main_html_v4
```

```python
import re

pattern = "\[\d+\]"

df['main_html_v3'] = df['main_html_v2'].apply(lambda x: re.sub(pattern, "", x))
```

```python
test_row = df.iloc[300]
main_html_v4 = test_row["main_html_v3"]
main_html_v4
```

```python
df['main_html_v4'] = df['main_html_v3'].apply(lambda x: BeautifulSoup(x).text)
df.head(13)
```

```python
df['main_html_v5'] = df['main_html_v4'].apply(lambda x: re.sub(r'http(\S+.*\s)', "", x))
df['main_html_v6'] = df['main_html_v5'].apply(lambda x: re.sub(r' (©2022|©2023|2022|2023)', "", x))
```

```python
test_row = df.iloc[3]
main_html_v4 = test_row["main_html_v6"]
main_html_v4
```

```python
df['main_html_v7'] = df['main_html_v6'].apply(lambda x: re.sub('\n', "", x))
df['main_html_v8'] = df['main_html_v7'].apply(lambda x: re.sub('\xa0', "", x))
```

```python
test_row = df.iloc[300]
main_html_v4 = test_row["main_html_v8"]
main_html_v4
```

```python
df2 = df.drop(["main_html_v3", "main_html_v4", "main_html_v5", "main_html_v6", "main_html_v7"], axis=1)
df2
```

```python
df2.to_csv(f"{OUTPUT_FOLDER}/{OUTPUT_DATA_FILE}", sep=";", index=False)
```
